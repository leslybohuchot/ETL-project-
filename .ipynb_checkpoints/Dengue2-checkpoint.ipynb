```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import pymongo
from bs4 import BeautifulSoup
warnings.filterwarnings("ignore")
%matplotlib inline

#bring in your data
data = "./dengue_prediction/data/dengue_features_train.csv"
data_2 = pd.read_csv("./dengue_prediction/data/dengue_labels_train.csv")
rain = pd.read_csv(data)
rain.head()


```

```python
#establish a connection and start a connection 
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.Wikipedia_db
collection = db.items
```

```python
#plot the basic data on year and total dengue cases for PR

sns.distplot(data_2.year)

year = rain[rain.columns[1:9]]
year.head()
year.plot()
plt.show()

sns.distplot(data_2.total_cases)
plt.show()
```

```python
#import url to compare total dengue cases with the rest of the world
url = "https://en.wikipedia.org/wiki/Dengue_fever_outbreaks"
# import requests
# response = requests.get(url)
# soup=BeautifulSoup(response.text, 'lxml')
# soup.find_all()

#read the html file
dengue = pd.read_html(url)
df = dengue[0]
df.columns = df.iloc[0]
df = df.reindex(df.index.drop(0))

#transform into a dictionary
dic = df.to_dict(orient="records")
dic
```

```python
#show the data frame to compare data
df
```

```python

#transfer the dictionary into mongo DB
collection.insert_many(dic)
```
